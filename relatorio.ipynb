# Prática 6: Algoritmos Genéticos
Na prática a seguir implementamos um algoritmo genético para encontrar os melhores valores de x e y para encontrar os mínimos da função de teste de otimização de mishra, a bird function.

Inicialmente vamos instanciar a classe criada

```python
from genetic import GeneticAlgorithym
import matplotlib.pyplot as plt
```

```python
#inicialização da classe Genetic
ga = GeneticAlgorithym()
```

Em seguida, iremos treinar 100 gerações de 100 indivíduos com uma taxa de cruzamento de 70% e uma taxa de mutação de 1%

```python
worse_fit,best_fit,avg_fit,best_individual= ga.run(sample_size=100,n_iterations=100,crossing_rate=0.7,mutation_rate=0.01)
print(f"Melhor indivíduo encontrado:  {best_individual}")
plt.plot(worse_fit,'r')
plt.plot(best_fit,'g')
plt.plot(avg_fit,'b')
plt.show()
```

Como esperado, os indivíduos convergiram para um resultado próximo ao ótimo. O pior indivíduo de cada geração manteve-se mais alto que a média pois, provavelmente, foi um indivíduo mutado que divergiu do valor ótimo.

O resultado obtido foi esperado, obtendo-se uma média de 95-99% de acurácia nos resultados.
